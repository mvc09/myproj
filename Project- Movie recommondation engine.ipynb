## Inspection

```python
df = df[['color','movie_title','gross','genres','director_name','director_facebook_likes','actor_1_name',
        'actor_1_facebook_likes', 'actor_2_name', 'actor_2_facebook_likes','actor_3_name',
        'actor_3_facebook_likes','cast_total_facebook_likes','movie_facebook_likes','num_critic_for_reviews','num_voted_users',
        'num_user_for_reviews','facenumber_in_poster','plot_keywords','language','country',
         'content_rating','budget','title_year','imdb_score','movie_imdb_link','aspect_ratio',]]
df.head()
```

```python
df.isnull().sum()  # Column
```

```python
df.isnull().sum(axis=1) # rows
```

```python
df[["1","2","3","4", "5", "6","7","8"]] = df["genres"].str.split("|", expand=True)
df
```

```python
df.drop(["movie_imdb_link","aspect_ratio", "actor_2_name","actor_2_facebook_likes",
"actor_3_name","actor_3_facebook_likes","2","3","4", "5", "6","7","8"],axis=1)
```

```python
df3 = df.groupby(["1"])["gross"].sum()
type(df3)
df3
```

```python
a,b,c = plt.pie(df3, radius=2,labels=df3.index, pctdistance=1,autopct='%.2f')
plt.legend(loc='upper right', bbox_to_anchor=(-1, 1.),
           fontsize=8)
```

```python
bardf = df.groupby("1")["gross"].mean()
plt.bar(bardf.index, bardf)
a = plt.xticks(bardf.index, rotation=45,horizontalalignment="right")
```

```python
z = pd.concat([df["gross"]/10000, df["budget"]/1000000])
```

```python
plt.plot(z)
plt.xlabel("gross/10k")
a =plt.ylabel("budget/1000k")
```
