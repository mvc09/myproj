```python
df1 = df.dropna(subset=["genres", "movie_title","language"]).reset_index(drop=True)
```

```python
df1[["1", "2", "3", "4","5","6","7","8"]] = df1["genres"].str.split("|", expand=True)
df1
```

```python
df1.drop(["2", "3", "4","5","6","7","8"], axis=1)
```

```python
df3 = df1.groupby(["1"])["gross"].sum()
type(df3)
df3
```

```python
a,b,c = plt.pie(df3, radius=2,labels=df3.index, pctdistance=1,autopct='%.2f')
plt.legend(loc='upper right', bbox_to_anchor=(-1, 1.),
           fontsize=8)
```

```python
x = pd.Series(df1["gross"])
y = pd.Series(df1["budget"])
z = pd.concat([y/100000,x/1000000], axis=1)
z.dropna().reset_index(drop=True)
z.set_index("budget", inplace=True)
```

```python
plt.plot(z,"ro")
plt.xlabel("budget")
plt.ylabel("gross")
```
